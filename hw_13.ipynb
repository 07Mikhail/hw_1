#### сколько лошадей с некорректным значением по surgery?

#### по возрасту?

```python
df[df.Age > 2].count() # по условию значения от 0 до 2
```

```python
df[df.Age < 0].count() # по условию значения от 0 до 2
```

```python
def my_function(df):
    """
    заполним ненормальные (24) значения, в зависимости от среднего значения пульса
    """
    if df['Age'] > 2:
        if df['pulse'] > df['pulse'].mean():
            return 2
        else:
            return 1
    else:
        return df['Age']

df['new_age'] = df.apply(my_function, axis=1)
del df['Age']
df[df.new_age > 2].count()
```

# 5. Проверяем выбросы по всем стобцам - удалить или изменить?

#### по surgery

```python
q1 = df['surgery?'].quantile(0.25)
q3 = df['surgery?'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
result = df[df['surgery?'].between(lower_bound, upper_bound, inclusive=True)].sort_values('surgery?')
result
```

```python
# что это за выброс?
pd.concat([df, result]).drop_duplicates(keep=False)
```

```python
q1 = df['pulse'].quantile(0.25)
q3 = df['pulse'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
result = df[df['pulse'].between(lower_bound, upper_bound, inclusive=True)].sort_values('pulse')
result
```

```python
def my_function(df):
    """
    заменим текущие 5 значений выбросов - на среднее арифметическое
    """
    if df['pulse'] >= 150:
        return df['pulse'].mean()
    else:
        return df['pulse']

df['new_pulse'] = df.apply(my_function, axis=1)
del df['pulse']
df[df.new_pulse > 150].count()
```

```python
df.head(10).sort_values('new_pulse', ascending=False) ### почему не заменилось 164?
```

#### по respiratory rate

```python
q1 = df['respiratory rate'].quantile(0.25)
q3 = df['respiratory rate'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
result = df[df['respiratory rate'].between(lower_bound, upper_bound, inclusive=True)].sort_values('respiratory rate')
result
```

```python
# что это за выброс?
pd.concat([df, result]).drop_duplicates(keep=False).sort_values('respiratory rate')
```

```python
def my_function(df):
    """
    т.к. полезность сомнительна из-за больших колебаний, заменим - на среднее арифметическое
    """
    if df['respiratory rate'] >= 66:
        return df['respiratory rate'].mean()
    else:
        return df['respiratory rate']

df['new_respiratory rate'] = df.apply(my_function, axis=1)
del df['respiratory rate']
df[df['new_respiratory rate'] > 66].count()
```
