```python
url_3 = 'C:/Users/vokea/OneDrive/Рабочий стол/Mishin/ml-latest-small/Новая папка/ml-latest-small/ratings.csv'
data_ = pd.read_csv(url_3, sep=',')
data_.head(1)
```

```python
data_['date'] = pd.to_datetime(data_['timestamp'], unit='s')
data_.head(1)
```

```python
data_count = data_.groupby(['userId']).count()
data_count.head(1)
```

```python
data_aggr = data_.groupby(['userId']).agg([min, max])
data_aggr['Разница'] = data_aggr.date['max'] - data_aggr.date['min']
data_aggr.head(1)
```

```python
del data_aggr['rating']
data_aggr.head(1)
```

```python
del data_aggr['movieId']
data_aggr.head(1)
```

```python
del data_aggr['date']
data_aggr.head(1)
```

```python
del data_aggr['timestamp']
data_aggr.head(1) ### готовые очищенные данные
```

```python
data_filter = data_count.loc[data_count.movieId > 100]
data_filter.head(1)
```

```python
f_join = data_filter.join(data_aggr, rsuffix='left')
f_join.head(5)
```

```python
del f_join['rating']
f_join.head(1)
```

```python
del f_join['timestamp']
f_join.head(1)
```

```python
del f_join['date']
f_join.head(1)
```

```python
f_join.rename(columns={'movieId': 'Кол-во оценок', '(Разница, )': 'Время жизни'}, inplace=True)
f_join.head(10)
```
